```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data=pd.read_csv('winequality.csv')
print(data.head())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(x='type',data=data)
plt.title('Occurrences of Wine Types')
plt.show()
```

```python
sns.countplot(x='quality',data=data)
plt.title('Count Of Quality Wines')
plt.show()
```

```python
sns.countplot(x='quality',data=data,hue='type')
plt.title('Count Of Quality Wine Types')
plt.show()
```

```python
data.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
sns.scatterplot(x='quality',y='alcohol',data=data,hue='type')
plt.title('Quality Vs Alcohol %')
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.scatterplot(y='residual sugar',x='alcohol',data=data,hue='type')
plt.title('alcohol % vs residual sugar')

plt.subplot(2,2,2)
sns.scatterplot(y='pH',x='alcohol',data=data,hue='type')
plt.title('alcohol % vs pH value')

plt.subplot(2,2,3)
sns.scatterplot(y='citric acid',x='alcohol',data=data,hue='type')
plt.title('alcohol % vs citric acid')

plt.subplot(2,2,4)
sns.scatterplot(y='density',x='alcohol',data=data,hue='type')
plt.title('alcohol % Vs residual sugar')

plt.show()
```

```python
sns.violinplot(x='quality',y='density',data=data)
plt.title('Quality Vs Density')
plt.show()
```

```python
sns.violinplot(x='quality',y='fixed acidity',data=data)
plt.title('Quality Vs Fixed Acidity')
plt.show()
```

```python
sns.violinplot(x='quality',y='volatile acidity',data=data)
plt.title('Quality Vs Volatile Acidity')
plt.show()
```

```python
sns.violinplot(x='quality',y='sulphates',data=data)
plt.title('Quality Vs Sulphates')
plt.show()
```

```python
plt.figure(figsize=(11,11))

plt.subplot(2,2,1)
sns.violinplot(x='quality',y='density',data=data,hue='type',inner=None)
plt.title('Quality Vs Density')

plt.subplot(2,2,2)
sns.violinplot(x='quality',y='fixed acidity',data=data,hue='type',inner=None)
plt.title('Quality Vs Fixed Acidity')

plt.subplot(2,2,3)
sns.violinplot(x='quality',y='volatile acidity',data=data,hue='type',inner=None)
plt.title('Quality Vs Volatile Acidity')

plt.subplot(2,2,4)
sns.violinplot(x='quality',y='sulphates',data=data,hue='type',inner=None)
plt.title('Quality Vs Sulphates')

plt.show()
```

```python
plt.figure(figsize=(12,8))
df = data.drop(columns=['type'])
sns.heatmap(data=df.corr(),cmap='Blues',annot=True,fmt=".2f")
plt.title('Correlation Heatmap (Wine Features)')
plt.show()
```
